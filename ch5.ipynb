## Pivots and Reshaping Data

```python
import numpy as np
import pandas as pd
#import pandas_datareader as pdr
```

```python
plantGrowthRawDF=pd.read_csv('./Files/PlantGrowth.csv')       
plantGrowthRawDF 
```

```python
pgDFpivot = plantGrowthRawDF.pivot(index='observation',columns='group',values='weight')
pgDFpivot
```

```python
plantGrowthStackedDF = plantGrowthRawDF.set_index(['group','observation'])
plantGrowthStackedDF
```

```python
plantGrowthStackedDF.columns
```

```python
plantGrowthStackedDF.unstack(level='group')
```

```python
plantGrowthStackedDF.unstack(level='group').columns
```

```python
plantGrowthStackedDF.unstack()
```

```python
plantGrowthStackedDF.unstack(level=1)
```
